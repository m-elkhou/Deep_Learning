3.On  considère  un  réseau  de  type Adaline.  On  veut  apprendre  l’ensemble  d'associations  entre  les données et les classes. L’algorithme d’apprentissage est le suivant (voir le cours) :

4.Ecrire  une  fonction err_adal permettant  de  calculer  l’écart  entre  la  sortie  désirée  (classe)  et  la sortie calculée par Adaline.

```python
def err_adal(y, dk):
    return y-dk
```

5.Ecrire une fonction grad_adal permettant de calculer une approximation du gradient.

```python
def grad_adal(ek, X):
    return -2*X.T.dot(ek)
```

6.Ecrire  une  fonction adapt_adal permettant  de  mettre  à  jour  les  poids  des connexions  dans Adaline en fonction du gradient.

```python
def adapt_adal(w, eps, app_gar):
    return w - eps*app_gar
```

7.Ecrire   une   fonction Adaline utilisant   les   fonctions   précédentes   permettant   de   faire   un apprentissage du modèle Adaline sur la base de données générée précédemment. On utilisera la fonction de transition indiquée dans la figure ci-dessous, ainsi que les valeurs initiales des poids et un pas du gradient de ε=0.1.

```python
def Adaline(X, y, eps, itr):
#     w = np.random.rand(np.shape(X)[1] + 1)
    w = np.array([0.8, 0.4, 0.3])
    x = np.array([np.append(x, 1) for x in X])
    for i in range(len(y)):
        if y[i]==0:
            y[i]=-1
    all_w = []
    c=0
    while True:
        dkp = np.dot(x, w)
        dk = np.where(dkp > 0.0, 1, -1)
        ek = err_adal(y, dk)
        app_gar = grad_adal(ek, x)
        c+=1
        all_w.append(w)
        if c == itr:
            break
        w = adapt_adal(w, eps, app_gar)
    return w, dk, ek, c, all_w, app_gar
```

```python
w, dk, ek, c, all_w, app_ = Adaline(X, y, 0.1, 10)
```

```python
def plot_iters(X, y, all_w):
    x = np.linspace(np.amin(X[:,:1]),np.amax(X[:,:1]))
    for i in range(len(all_w)):
        plt.figure()
        fw = all_w[i]
        a = -(fw[2]/fw[1])/(fw[2]/fw[0])  
        b = -fw[2]/fw[1]
    
        plt.rcParams["figure.figsize"] = (15,5)
        plt.title('iteration '+str(i))
        plt.plot(x, [a*i + b for i in x], color='k')
        plt.scatter(X[y==-1, 0], X[y==-1, 1], label='class -1', marker='+')
        plt.scatter(X[y==1, 0], X[y==1, 1], label='class 1', marker='o')
        plt.xlabel('area')
        plt.ylabel('perimeter')
```

```python
plot_iters(X, y, all_w)
```

12.Ecrire  une  fonction <b>perceptron</b> utilisant  les  fonctions  précédentes  permettant  de  faire un apprentissage  du  modèle Perceptron sur  la  même  base  de  données  qu’Adaline.  On  utilisera les valeurs d’initialisation données dans la question 7.

13.A   chaque   itération   de   l’apprentissage   représenter   graphiquement   dans   un   même   repère orthogonal l’ensemble des échantillons et l’hyperplan séparant les deux classes.

```python
class Perceptron():
    def __init__(self, eps, w, itr):
        self.eps_ = eps
        self.w = w
        self.itr = itr
    
    def sortie_perc(self, X):
        return np.where((X.dot(self.w[1:])+self.w[0])>0.0,1,-1)
    
    def adapt_perc(self, y, X):
        self.w[1:] += self.eps_*(y - self.py)*X
        self.w[0] += self.eps_*(y - self.py)*self.w[0]
        return self.w
    
    def fit(self, X, y):
        for _ in range(self.itr):
            for xi, yi in zip(X, y):
                self.py = self.sortie_perc(xi)
                self.w = self.adapt_perc(yi, xi)
    
    def predict(self, X):
        return self.sortie_perc(X)
    
```

```python
p_w = np.array([0.3, 0.8, 0.4])
per = Perceptron(0.1, p_w, 15)
per.fit(X,y)
py = per.predict(X)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(X[:,0], X[:,1], c=y)
plt.title('les résultats désirés')

plt.subplot(1,2,2)
plt.scatter(X[:,0], X[:,1], c=py)
plt.title('les résultats obtenus')

plt.show()
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X, py, clf=per)
```
