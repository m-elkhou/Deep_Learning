<b> A) Régression avec un Perceptron Multi-Couches</b>

1.Créez  un  ensemble  de  points  (une 20ène)  suivant  une  fonction  sinusoïde  avec  un  bruit gaussien d'écart type 0,2.
On va essayer de retrouver la fonction à partir des points avec un  Perceptron  Multi-Couches  (MLP).
Analyser  la  documentation  sur  MLP  dans  la librairie scikit-learn et analysez toutes les parametres.

x = np.linspace(0,8,20).reshape(-1,1)

y = np.sin(x) + np.random.randn(20)*0.2

```python
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.linspace(0,8,20).reshape(-1,1)
y = np.sin(x) + np.random.randn(20)*0.2
```

2.Initialisez  le  MLP  avec  la  fonction MLPRegressor.  On  veut  3  neurones  cachés,  une fonction  d'activation  linéaire  et  un  pas  d'apprentissage  (learning  rate)  de  0,1  avec  un nombre d’iterations de 100.

```python
from sklearn.neural_network import MLPRegressor

MLP = MLPRegressor(hidden_layer_sizes=3, activation='logistic', learning_rate_init=0.1, max_iter=300)
```

3.Lancez l'apprentissage du MLP. Utilisez la fonction fit.

```python
MLP.fit(x,y[:,1])
```

4.La fonction predictpermet d'utiliser le MLP appris pour faire de la régression.

```python
y_p = MLP.predict(x)
```

5.Tracez  sur  une  même  figure  les  points  d'apprentissage,  la  fonction  sinusoïde  réelle  et celle estimée par le MLP.

```python
plt.figure()
plt.plot(x,y[:,1], c='b', label='true y')
plt.plot(x,y_p, c='r', label='y predict')
plt.legend(loc='best')
plt.xlabel('les points d\'apprentissage, la fonction sinusoïde réelle et celle estimée par le MLP.')
plt.show()
```

<b>B)Classification avec un Perceptron Multi-Couches</b>

1.Les  données  IRIS  sont constituées  de  4  colonnes  qui  correspondent  à  4  variables  de description. La 5ème colonne donne la classe de chaque objet. Transformez ces données en deux matrices contenant les données et le label (target).

```python
import sklearn.datasets as sd

iris = sd.load_iris()
data = iris.data
label = iris.target
```

2.Pour permettre l'apprentissage par un MLP, il faut «binariser» les numéros de classe.

3.Initialisez  le  MLP  avec  la  fonction MLPClassifier.  On  veut  2  neurones  cachés,  une fonction d'activation logistique et un pas d'apprentissage de 0,2.

```python
from sklearn.neural_network import MLPClassifier

MLPC = MLPClassifier(hidden_layer_sizes=2, activation='logistic', learning_rate_init=0.2, solver='lbfgs')
```

4.On  garde  les  mêmes  options  que  dans  l'exercice  precedent.  Lancez l'apprentissage  du MLP  avec  la  fonction fit.  On  veut  utiliser  l'algorithme  d'optimisation de  descente  de gradient stochastique (lbfgs).

```python
MLPC.fit(data, label)
```

5.Affichez les résultats désirés et les résultats obtenus dans une même figure. Vous pouvez utiliser  un  ACP  par  exemple. Chaque  point  aura  une  couleur  différente  selon  sa  classe. Les résultats sont-ils bon?

```python
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score

pca = PCA(n_components=2)
pca_pr = pca.fit(data).transform(data)

mlp_pr = MLPC.predict(data)
print('accuracy_score : ', accuracy_score(label, mlp_pr))

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(pca_pr[:,0], pca_pr[:,1], c=label)
plt.title('les résultats désirés')

plt.subplot(1,2,2)
plt.scatter(pca_pr[:,0], pca_pr[:,1], c=mlp_pr)
plt.title('les résultats obtenus')

plt.show()
```

6.Essayez   de   normaliser   les   données   (centrées –réduites).   Cela   améliore-t-il   la classification?

```python
from sklearn.preprocessing import scale

data = scale(data)
MLPC.fit(data, label)
mlp_pr = MLPC.predict(data)

mlp_pr = MLPC.predict(data)
print('accuracy_score : ', accuracy_score(label, mlp_pr))

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(pca_pr[:,0], pca_pr[:,1], c=label)
plt.title('les résultats désirés')

plt.subplot(1,2,2)
plt.scatter(pca_pr[:,0], pca_pr[:,1], c=mlp_pr)
plt.title('les résultats obtenus')

plt.show()
```

7.Essayez  de  faire  varier  les  paramètres  du  MLP  (nombre  de  neurones  cachés,  etc...). Comment ces paramètres influencent les résultats?

```python
MLPC = MLPClassifier(hidden_layer_sizes=5, activation='logistic', learning_rate_init=0.1, solver='lbfgs')
MLPC.fit(data, label)
mlp_pr = MLPC.predict(data)
print('accuracy_score : ', accuracy_score(label, mlp_pr))

plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.scatter(pca_pr[:,0], pca_pr[:,1], c=label)
plt.title('les résultats désirés')

plt.subplot(1,2,2)
plt.scatter(pca_pr[:,0], pca_pr[:,1], c=mlp_pr)
plt.title('les résultats obtenus')

plt.show()
```

<b>C)Réduction de dimensions avec un Perceptron Multi-Couches</b>

1.Entraînez un MLP à retrouver les vecteurs d'entrée en sortie (la sortie doit être la même que  l'entrée) sur  la  base  de  données  Iris.  On  utilisera  ici  deux  neurones cachés  et  une fonction d'activation linéaire.

```python
MLPC = MLPClassifier(hidden_layer_sizes=2, activation='relu', learning_rate_init=0.1, solver='lbfgs')
MLPC.fit(data, label)
mlp_pr = MLPC.predict(data)
print('accuracy_score : ', accuracy_score(label, mlp_pr))
```

2.Après  apprentissage,  calculez  les  sorties  de  la  couche  cachée  pour  chaque  point  de données  en  utilisant  la  matrice  de  poids(coef) et  les  données  en  entrées. Vous  obtenez deux valeurs  par  points,  qui  donnent  de  nouvelles  coordonnées  pour  chaque  point. Affichez ces valeurs dans une figure (avec les classes réelles représentées en couleur).

```python
w1=MLPC.coefs_[0]
c1=data.dot(w1)
plt.scatter(c1[:,0], c1[:,1], c=label)
plt.show()
```
